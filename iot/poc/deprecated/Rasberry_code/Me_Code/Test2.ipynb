```python
import paho.mqtt.client as mqtt

print("Version de la bibliothèque Paho MQTT : ", mqtt.__version__)

def on_connect(client, userdata, flags, rc):
    print("Connecté avec le code de retour "+str(rc))
    client.subscribe("topic/path")

def on_message(client, userdata, msg):
    print(msg.topic+" "+str(msg.payload))
    # Traitez les données reçues ici et générez une réponse
    response = generate_response(msg.payload)
    # Envoyez la réponse au topic "response"
    client.publish("response", response)

def generate_response(data):
    # Traitez les données reçues et générez une réponse
    return "Réponse à " + str(data)

def create_mqtt_server(host, port):
    client = mqtt.Client(protocol=mqtt.MQTTv311)  # Spécifiez la version de l'API de rappel
    client.on_connect = on_connect
    client.on_message = on_message

    client.connect(host, port, 60)

    # Attend indéfiniment
    client.loop_forever()  # Démarre la boucle de gestion MQTT

if __name__ == "__main__":
    HOST = '0.0.0.0'  
    PORT = 1883        
    create_mqtt_server(HOST, PORT)
```
