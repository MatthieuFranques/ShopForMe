Function Ping 

```python
import socket

def ping(host, port):
    # Créer un socket TCP/IP
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        try:
            # Essayer de se connecter au serveur
            s.connect((host, port))
            return True
        except Exception as e:
            # En cas d'erreur de connexion, renvoyer False
            print(f"Erreur de connexion : {e}")
            return False

# Adresse IP et port à ping
host = '192.168.26.1'
port = 1883

# Appeler la fonction ping
if ping(host, port):
    print(f"Le serveur {host} sur le port {port} est accessible.")
else:
    print(f"Le serveur {host} sur le port {port} est inaccessible.")
```

Function open port 

```python
import socket

def ouvrir_port(port):
    # Création d'un objet de type socket
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    
    try:
        # Liaison du socket à l'adresse IP de votre PC et au port spécifié
        s.bind(('0.0.0.0', port))
        print("Port {} ouvert avec succès.".format(port))
    except Exception as e:
        print("Erreur lors de l'ouverture du port {}: {}".format(port, e))
    finally:
        # Fermeture du socket
        s.close()

# Port que vous souhaitez ouvrir
port_a_ouvrir = 1883

# Appel de la fonction pour ouvrir le port
ouvrir_port(port_a_ouvrir)
```
