- Pr :  est la puissance reçue (en dBm).
- Pt :  est la puissance transmise (en dBm).
- Gt :  est le gain de l'antenne de transmission (en dBi).
- Gr :  est le gain de l'antenne de réception (en dBi)
- λ  :  est la longueur d'onde du signal (en mètres).
- d  :  est la distance entre les deux antennes (en mètres).

Source : https://fr.wikipedia.org/wiki/Équation_des_télécommunications

```python
import math

def calculate_rssi(Pt_dBm, Gt_dBi, Gr_dBi, frequency_hz, distance_m):

    """
  La fonction calcule la longueur d'onde à partir de la fréquence,
  puis utilise l'équation de Friis pour déterminer le RSSI à la distance spécifiée.
  Enfin, elle retourne le RSSI en dBm.

    Args:
    - Pt_dBm :          Puissance d'émission en dBm.
    - Gt_dBi :          Gain de l'antenne de transmission en dBi.
    - Gr_dBi :          Gain de l'antenne de réception en dBi.
    - frequency_hz :    Fréquence du signal en Hz.
    - distance_m  :     Distance entre les antennes en mètres.
    
    Returns:
    - float: RSSI en dBm.
    """
    # Constantes
    c = 3e8  # Vitesse de la lumière en m/s

    # Calcul de la longueur d'onde
    wavelength_m = c / frequency_hz

    # Calcul du terme de propagation
    propagation_loss = 20 * math.log10(wavelength_m / (4 * math.pi * distance_m))

    # Calcul du RSSI
    Pr_dBm = Pt_dBm + Gt_dBi + Gr_dBi + propagation_loss

    return Pr_dBm

# Exemple d'utilisation
Pt_dBm = 20  # Puissance d'émission en dBm
Gt_dBi = 2   # Gain de l'antenne de transmission en dBi
Gr_dBi = 2   # Gain de l'antenne de réception en dBi
frequency_hz = 2.4e9  # Fréquence en Hz (ex. 2.4 GHz pour Wi-Fi)
distance_m = 3  # Distance en mètres

rssi = calculate_rssi(Pt_dBm, Gt_dBi, Gr_dBi, frequency_hz, distance_m)
print(f"Le RSSI à {distance_m} mètres est de {rssi:.2f} dBm")
```

- frequency_hz : Fréquence du signal en Hz (2.4 GHz pour Wi-Fi = 2.4e9 Hz).

```python
import math

def calculate_distance(Pt_dBm, Gt_dBi, Gr_dBi, frequency_hz, Pr_dBm):
    # Constantes
    c = 3e8  # Vitesse de la lumière en m/s

    # Calcul de la longueur d'onde
    wavelength_m = c / frequency_hz

    # Calcul de la distance
    distance_m = (wavelength_m / (4 * math.pi)) * 10 ** ((Pt_dBm + Gt_dBi + Gr_dBi - Pr_dBm) / 20)

    return distance_m

# Exemple d'utilisation
Pt_dBm = 20  # Puissance d'émission en dBm
Gt_dBi = 2   # Gain de l'antenne de transmission en dBi
Gr_dBi = 2   # Gain de l'antenne de réception en dBi
frequency_hz = 2.4e9  # Fréquence en Hz (ex. 2.4 GHz pour Wi-Fi)
Pr_dBm = -22.46  # RSSI mesuré en dBm
distance = calculate_distance(Pt_dBm, Gt_dBi, Gr_dBi, frequency_hz, Pr_dBm)
print(f"La distance estimée est de {distance:.2f} mètres")
```
